```python
import pandas as pd
```

```python
new_out = open('Output/Morex/new_intragenic_context_intron_exon_IL_IP_IR.tsv','a')
with open('Output/Morex/intragenic_context_intron_exon_IL_IP_IR.tsv') as f:
    for line in f:
        line = line.rstrip()
        line = line.split()
        inex_id = line[6]
        inex_type = line[7]
        if  inex_type == 'intron':
            new_inex_id = 'intron.' + inex_id.split('.')[6]
            new_out.write('\t'.join(line[:6]) + '\t' + new_inex_id + '\t' + '\t'.join(line[7:])+'\n')
        else:
            new_inex_id = '.'.join(inex_id.split('.')[5:])
            new_out.write('\t'.join(line[0:6])+ '\t' + new_inex_id + '\t'+ '\t'.join(line[7:])+ '\n')
            
new_out.close()            
```

```python
new_out_2 = open('Output/Morex/new_intragenic_context_InEx_IL_IP_IR.tsv','a')
with open('Output/Morex/intragenic_context_InEx_IL_IP_IR.tsv') as f:
    for line in f:
        line = line.rstrip()
        line = line.split()
        inex_id = line[6]
        inex_type = line[7]
        new_inex_id = '.'.join(inex_id.split('.')[5:])
        new_out_2.write('\t'.join(line[0:6])+ '\t' + new_inex_id + '\t'+ '\t'.join(line[7:])+ '\n')           
new_out_2.close()     
```

```python
df_Intron_Exon = pd.read_csv('Output/Morex/new_intragenic_context_intron_exon_IL_IP_IR.tsv', sep = '\t', names = ['chr','start','end','type','cultivar','gene_ID','InEx_ID','InEx_type','gene_type'])
df_Intron_Exon
```

```python
df_InEx = pd.read_csv('Output/Morex/new_intragenic_context_InEx_IL_IP_IR.tsv', sep = '\t', names = ['chr','start','end','type','cultivar','gene_ID','InEx_ID','InEx_type','gene_type'])
df_InEx
```

```python
df_tot = pd.concat([df_InEx,df_Intron_Exon])
df_tot
```

```python
df_list = [x[1] for x in df_tot.groupby('cultivar', sort=True)]
```

```python
for i in range(len(df_list)):
    tot = sum(df_list[i].value_counts(df_list[i]['gene_type']))
    print('')
    print('Cultivar: '+(df_list[i]['cultivar'][:1]).to_string(index=False))
    print('')
    print('###'+(df_list[i].value_counts(df_list[i]['gene_type'])).to_string() + '\n\n###Percentage ' + (df_list[i].value_counts(df_list[i]['gene_type'])/tot).to_string())
    print('')
    print('###'+(df_list[i].value_counts(df_list[i]['InEx_type'])).to_string() + '\n\n###Percentage ' + (df_list[i].value_counts(df_list[i]['InEx_type'])/tot).to_string())
    print('')    
    print('###'+(df_list[i].value_counts(df_list[i]['InEx_ID']).nlargest(5)).to_string() + '\n\n###Percentage ' + ((df_list[i].value_counts(df_list[i]['InEx_ID']).nlargest(5))/tot).to_string())
    print('')
    print('###'+(df_list[i].value_counts(df_list[i]['gene_ID']).nlargest(5)).to_string() + '\n\n###Percentage ' + ((df_list[i].value_counts(df_list[i]['gene_ID']).nlargest(5))/tot).to_string())
    print('------------------------------')
```

```python
df_list_gene_based = [x[1] for x in df_tot.groupby('gene_ID', sort=True)]
```

```python
out_file_gene = open('Statistics_gene_based.txt','a')
for i in range(len(df_list_gene_based)):
    tot = sum(df_list_gene_based[i].value_counts(df_list_gene_based[i]['gene_type']))
    out_file_gene.write('')
    out_file_gene.write('Gene ID: '+(df_list_gene_based[i]['gene_ID'][:1]).to_string(index=False))
    out_file_gene.write('')
    out_file_gene.write('\n###'+(df_list_gene_based[i].value_counts(df_list_gene_based[i]['gene_type'])).to_string() + '\n\n###Percentage ' + (df_list_gene_based[i].value_counts(df_list_gene_based[i]['gene_type'])/tot).to_string())
    out_file_gene.write('')
    out_file_gene.write('\n###'+(df_list_gene_based[i].value_counts(df_list_gene_based[i]['InEx_type'])).to_string() + '\n\n###Percentage ' + (df_list_gene_based[i].value_counts(df_list_gene_based[i]['InEx_type'])/tot).to_string())
    out_file_gene.write('')    
    out_file_gene.write('\n###'+(df_list_gene_based[i].value_counts(df_list_gene_based[i]['InEx_ID']).nlargest(5)).to_string() + '\n\n###Percentage ' + ((df_list_gene_based[i].value_counts(df_list_gene_based[i]['InEx_ID']).nlargest(5))/tot).to_string())
    out_file_gene.write('')
    out_file_gene.write('\n###'+(df_list_gene_based[i].value_counts(df_list_gene_based[i]['cultivar']).nlargest(5)).to_string() + '\n\n###Percentage ' + ((df_list_gene_based[i].value_counts(df_list_gene_based[i]['cultivar']).nlargest(5))/tot).to_string())
    out_file_gene.write('\n------------------------------\n')
out_file_gene.close()
```
